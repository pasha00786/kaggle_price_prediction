```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv('num_train.csv')
data.drop(['Unnamed: 0'], axis = 1, inplace = True)
data.head()
```

```python
X = np.array(data.drop(['SalePrice'], axis = 1))
y = np.array(data.SalePrice).reshape(-1,1)
# mean_X = np.mean(X)
# std_X = np.std(X)
# mean_y = np.mean(y)
# std_y = np.std(y)
# X = (X - mean_X)/std_X
# y = (y - mean_y)/std_y
```

```python
#from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_log_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 16)
```

```python
#model = Lasso(alpha = 400.0, normalize = True)
model = Ridge(alpha = 1.9, normalize = True)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print((y_pred[59]), (y_test[59]))
```

```python
np.sqrt(mean_squared_log_error(y_test, y_pred))
```

```python
print( mean_squared_error(y_test, y_pred))
```
